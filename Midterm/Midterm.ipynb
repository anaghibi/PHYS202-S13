```python
y = lambda theta: (cos(2*theta)*cos(theta/2))**2
x = arange(-pi,pi,0.1)
```

```python
plot(x,y(x),color = "red")
plot(x,y(x),'b*')
xlabel(r'${\theta}$(rad)',fontsize=15)
ylabel('Intensity',fontsize=15)
xticks(arange(-pi,5*pi/4,pi/4), ['$-{\pi}$','$-3{\pi}/4$','$-{\pi}/2$','$-{\pi}/4$','0','${\pi}/4$','${\pi}/2$','$3{\pi}/4$','${\pi}$'])
```

###Part b)

```python
t = arange(-5,1,0.1)
f = lambda t:(t+1)/(t-1)
g = lambda t:e**(-t)
plot(f(t),t, label = '(t+1)/(t-1)')
plot(g(t),t,color='red',linestyle=':',label = 'exp(-t)')
xlim(-5,5)
ylim(-10,10)
legend(loc = 'lower left')
xlabel('time(seconds)')
ylabel('f(t),g(t)')
```
