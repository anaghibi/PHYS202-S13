```python
f1 = lambda x: a*(B**(-a))*(x**(a-1))*(e**(-(x/B)**(a)))
F = lambda x: 1-e**(-(x/B)**a) #CDF
Finv = lambda x: B*(-log(1-x))**(1/a) #inverse CDF

a = 1.5
B = 6
N = 10000 # 10,000 random samples are desired
x = arange(0,15,0.001)
#z = numpy.random.random_sample(N) #I can't figure out how to do this without using np.random
z = random_sample(N)
samples = Finv(z)

hist(samples,100,normed=True,label='Sampled distribution', color = 'y')
plot(x,f1(x),label='True distribution',lw=3,color='magenta')
plot(x,F(x),label='F',lw=3,color='c')
xlim(0,15)
ylim(0,1.1)
legend(loc=0)
show()
```
