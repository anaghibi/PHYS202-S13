```python
x = arange(0.,pi+0.001,0.01)
f = lambda x: sin(x)/2.
F = lambda x: (cos(0)-cos(x))/2.
               
plot(x,f(x),label='True Distribution',lw=3)
plot(x,F(x),label='Cumulative Function',lw=3)

xlim(0,3.5)
ylim(0,1.1)

legend(loc=0)
show()
```

```python
# number of samples
N = 10000

#N random numbers between 0 and 1
z = numpy.random.random_sample(N);

#Inverse CDF
Finv = lambda z: arccos(cos(0)-2*z)
    
# sampled x values
xsamples = Finv(z);

#how do they compare to the original?
hist(xsamples,50,normed=True,label='sampled distribution')
plot(x,f(x),label='True distribution',lw=3,color='red')

xlim(0,pi)
ylim(0,0.7)

legend(loc=0)
show()
```

```python
f = lambda x: 1*e**(-1*x)
x = arange(0.,5.+0.001,0.01)
F = cumsum(f(x))/sum(f(x)) #CDF

plot(x,f(x),label='True Distribution',lw=3)
plot(x,F,label='Cumulative Function',lw=3)

xlim(0,5.)
ylim(0,1.1)

legend(loc=0)
show()
```

###Part c)

```python
def distribute1D(x, prob, N):
    '''takes any numerical distribution prob, on the interval defined by the array x, ad returns an array of N sampled
    values that are statistically the same as the input data.'''
    distribution = prob(x)
    F = cumsum(distribution)/sum(distribution)
    z = numpy.random.random_sample(N);
    Finv = x[searchsorted(F,z)]
    return Finv
```

###Parts a&b)

```python
sinc = lambda x: sin(pi*x)/(pi*x)
Idiff = lambda x: sinc((a*x)/(lamb*L))**2
Iint = lambda x: cos((2*pi*d*x)/(lamb*L))**2
Itot = lambda x: Idiff(x) * Iint(x)

d = 15*(10**-6)
a = 10*(10**-6)
L = 1
lamb = 632.8*(10**-9)

N = 10000 #10,000 photons
x = arange(-0.2,0.201,.001) #.2 meters wide
dist = distribute1D(x,Itot,N)
n,bins,patches=hist(dist,250,normed=True,label='Sampled distribution')
plot(x,Itot(x)*max(n),'m',label = 'True distribution')
legend(loc=0)
xlim(-0.2,0.2)
```
