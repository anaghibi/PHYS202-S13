create a very simple plot

```python
t = arange(0.0, 1.0+0.01, 0.01)
c = cos(2*2*pi*t)
s = sin(4*pi*t)
```

```python
print pi
```

```python
plot(t, c, color = "blue",linestyle = ":", linewidth = 2.5, label = "$cos(4{\pi}t)$")
plot(t, s, color = "magenta", linewidth = 2.5, label = "$sin(4{\pi}t)$")
xlabel('time (s)')
ylabel('voltage (mV)')
#add title for convention
title('A Simple Plot', fontsize = 15)
grid(True)
legend(loc="best")
savefig('simple_plot.pdf')
```
