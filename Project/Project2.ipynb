```python
L = 40
n = 5
vmax = 10
x = np.random.randint(1,L,n)
x_poss = sort(x)
v_poss = np.random.randint(1,vmax,n)
g = np.zeros(n,dtype=int)
```

```python
def CA184(n,pos,vel,L,vmax):
    init_pos_car0 = pos[0] #Storing the initial position of the first car to use later
    for i in range(0,n):
        if i+1 < n: #If car is not the last one in the array
            if pos[i+1] > pos[i]: #if car is not the furthest along on the road
                g[i] = pos[i+1]-pos[i]
            else:
                g[i] = L - pos[i] + pos[i+1] #spaces if car is the furthest along on the road
        else: #if car is the last one in array
            if init_pos_car0 > pos[i]: #if the position of the first car > the position of the last spot
                g[i] = init_pos_car0 - pos[i]
            else:
                g[i] = L - pos[i] + init_pos_car0
        if  g[i] == 0: # car doesn't move if no space in front of it
            vel[i] = 0
        elif g[i] <= vel[i]: #increase v to g if there is space
            vel[i] = g[i]
        elif vel[i]<g[i] and vel[i]<10: #increase speed until max if there is space
            vel[i] = vel[i] + 1
            
    for i in range(0,n):
        if pos[i] + vel[i] <= L: #if car is not going off the road
            pos[i] = pos[i] + vel[i] #increase pos by the velocity
        else: #if car is going off the road
            pos[i] = pos[i] + vel[i] - L #inc pos by vel & loop around to the beginning of the road
    return pos,vel
```

```python
def STCA(n,pos,vel,Len,prob,vmax):
    init_pos_car0 = pos[0] #Storing the initial position of the first car to use later
    for i in range(0,n):
        if i+1 < n: #If car is not the last one in the array
            if pos[i+1] > pos[i]: #if car is not the furthest along on the road
                g[i] = pos[i+1]-pos[i]
            else:
                g[i] = L - pos[i] + pos[i+1] #spaces if car is the furthest along on the road
        else: #if car is the last one in array
            if init_pos_car0 > pos[i]: #if the position of the first car > the position of the last spot
                g[i] = init_pos_car0 - pos[i]
            else:
                g[i] = L - pos[i] + init_pos_car0
        if  g[i] == 0: # car doesn't move if no space in front of it
            vel[i] = 0
        elif g[i] <= vel[i]: #increase v to g if there is space
            vel[i] = g[i]
        elif vel[i]<g[i] and vel[i] < vmax: #increase speed until max if there is space
            vel[i] = vel[i] + 1
        elif np.random.random_sample(1) < prob and vel[i]>0:
            vel[i] = vel[i] - 1
            
    for i in range(0,n):
        if pos[i] + vel[i] <= L: #if car is not going off the road
            pos[i] = pos[i] + vel[i] #increase pos by the velocity
        else: #if car is going off the road
            pos[i] = pos[i] + vel[i] - L #inc pos by vel & loop around to the beginning of the road
    array = np.array([pos,vel])
    return array
```

```python
def ASEP(n,pos,vel,Len,vmax):
    random_array = np.random.permutation(n)
    init_pos_car0 = pos[0]
    for j in range(n):
        i = random_array[j]
        if i+1 < n: #If car is not the last one in the array
            if pos[i+1] > pos[i]: #if car is not the furthest along on the road
                g[i] = pos[i+1]-pos[i]
            else:
                g[i] = L - pos[i] + pos[i+1] #spaces if car is the furthest along on the road
        else: #if car is the last one in array
            if init_pos_car0 > pos[i]: #if the position of the first car > the position of the last spot
                g[i] = init_pos_car0 - pos[i]
            else:
                g[i] = L - pos[i] + init_pos_car0
        if  g[i] == 0: # car doesn't move if no space in front of it
            vel[i] = 0
        elif g[i] <= vel[i]: #increase v to g if there is space
            vel[i] = g[i]
        elif vel[i]<g[i] and vel[i]>0: #increase speed until max if there is space
            vel[i] = vel[i] + 1
        if pos[i] + vel[i] <= L: #if car is not going off the road
            pos[i] = pos[i] + vel[i] #increase pos by the velocity
        else: #if car is going off the road
            pos[i] = pos[i] + vel[i] - L #inc pos by vel & loop around to the beginning of the road
    return pos,vel
```
