##Plotting V(x,y)

$V(x,y) = \frac{kq}{\sqrt{x^{2}+(y-d)^{2}}} - \frac{kq}{\sqrt{x^{2}+(y+d)^{2}}}$

```python
dx = 0.06
dy = 0.06
x,y = meshgrid(arange(-2.,2.,dx),arange(-2.,2.,dy))
#making the function V(x,y)
k = 1; q = 1; d = 1
V=(k*q/(x**2+(y-d)**2)**(1/2.)) - (k*q/(x**2+(y+d)**2)**(1/2.))
```

```python
print V
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = figure()
ax = fig.gca(projection='3d')
p = ax.plot_surface(x, y, V, rstride=1, cstride=1, linewidth = 0, cmap=cm.jet_r)
cb = fig.colorbar(p, shrink=0.5)
```

```python
levels = array([8, 4, 2, 1, .5, .25, 0, -.25, -.5, -1, -2, -4, -8])
contour(x,y,V,levels,linewidths=4,cmap=cm.jet)
```
